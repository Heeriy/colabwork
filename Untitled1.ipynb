revenue from pizza sales

```python
# Revenue=Quantity*price
jointable=pd.merge(od,p,on='pizza_id')
jointable.head()
```

```python
total_revenue=jointable['quantity']*jointable['price']
total_revenue
```

```python
total_revenue.sum()
```

name of highest priced pizza

```python
highest_priced_pizza=p[p['price']==p['price'].max()]
highest_priced_pizza
```

most common pizza slice ordered

```python
cp=od.groupby('pizza_id')['quantity'].sum()
cp.idxmax()
```

list top five most ordered pizza along with their quality


```python
mergedf1=pd.merge(od,p,on='pizza_id')
mergedd2=pd.merge(mergedf1,pt,on='pizza_type_id')

cp2=mergedd2.groupby('name')['quantity'].sum()
cp2.sort_values(ascending=False).head()
```

list five least ordered pizza  along with their quality

```python
cp2.sort_values(ascending=True).head()
```
